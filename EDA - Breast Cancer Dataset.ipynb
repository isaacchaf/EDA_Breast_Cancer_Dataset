## EDA - Breast Cancer Dataset

```python
# Library Import
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Breast cancer datase
breast_cancer = datasets.load_breast_cancer()

# Pyhton Function
def type_cancer(value):
    cancer = breast_cancer.target_names[value]
    return cancer
```

```python
# Dataframe
cancer_data = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
cancer_data["TypeCancer"] = breast_cancer.target
# Editing column names
cancer_data.columns = cancer_data.columns.str.replace(" ", "")
cancer_data.TypeCancer = cancer_data.TypeCancer.apply(type_cancer)
cancer_data.head()
```

# Univariate Analysis

```python
# Data Size
cancer_shape = cancer_data.shape
print("Number of Observations: ", cancer_shape[0])
print("Number of Columns: ", cancer_shape[1])
```

```python
# Data Characteristics
cancer_data.describe().T
```

```python
# Data (Storage) Type
cancer_data.info()
```

```python
# Null values 
cancer_data.isnull().sum()
```

```python
# Variable Balance 
cancer_data.TypeCancer.value_counts()
```

```python
# Separating the datasets
mean_variables = cancer_data.filter(regex='^mean',axis=1)
mean_variables['Type'] = cancer_data['TypeCancer'].copy()

error_variables = cancer_data.filter('error$', axis=1)

worst_variables = cancer_data.filter(regex='^worst',axis=1)

mean_variables
```

```python
plt.figure(figsize = (100,100))
cancer_data.hist()
plt.suptitle("Histograms")
plt.show()

cancer_data.boxplot()
plt.title("Boxplot")
plt.show()
```

```python
for column in cancer_data.columns[:30]:
    g = sns.FacetGrid(cancer_data, hue = "TypeCancer", height = 4)
    g = g.map(sns.distplot, column)
    g = g.add_legend()
    plt.show()
```

```python
# Variable Means 
cancer_data.groupby(by = 'TypeCancer').mean().T
```

```python
cancer_data.groupby(by = 'TypeCancer').mean().plot(kind = "bar")
plt.title("Dataset Variables")
plt.ylabel("Features Mean")
plt.grid(True)
plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
```
